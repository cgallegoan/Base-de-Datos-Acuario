```
import collections
class DiGraph:
    white, gray, black = 0, 1, 2 # permite hacer DiGraph.white
    
    def __init__(self):
        # dict associating a list of adjacent edges to each vertex
        # each edge is a tuple (destiny, weight)
        self._adj = collections.OrderedDict() # (adj from adjacents)

    def number_of_nodes(self): # |V|
        return len(self._adj)

    def nodes(self): # para iterar sobre los vÃ©rtices
        return self._adj.keys()

    def out_degree(self, u):
        return len(self._adj[u])

    def in_degree(self, u):
        # VERY costly!
        sum(1 for (w,weight) in lst for lst in self._adj.values() if w == u)
    
    def number_of_edges(self): # |A|
        return sum(self.out_degree(u) for u in self.nodes())
    
    def add_node(self, u):
        if u not in self._adj:
            self._adj[u] = []

    def add_edge(self, u, v, weight=1):
        # let's assume there is no edge (u,v)
        self.add_node(u) # just in case
        self.add_node(v) # just in case
        self._adj[u].append((v,weight))

    def weight_edge(self, u, v):
        # returns None if there is no such edge (u,v)
        if u in self._adj:
            for w,weight in self._adj[u]:
                if v == w:
                    return weight

    def transpose(self):
        resul = DiGraph()
        for u in self.nodes():
            for v,w in self._adj[u]:
                resul.add_edge(v, u, w)
        return resul
        
    def BFS(self, src):
        colour = { u:DiGraph.white for u in self.nodes() }
        pred, dist = { src:None }, { src:0 }
        Q = collections.deque()
        colour[src] = DiGraph.gray
        Q.append(src)
        i = 0
        while len(Q) > 0:
            print(f'\niter {i}\ncolour {colour}\npred {pred}\ndist {dist}\nQ {Q}')
            i += 1
            u = Q.popleft()
            print(f'se selecciona {u}')
            for (v, weight) in self._adj[u]:
                if colour[v] == DiGraph.white:
                    dist[v] = dist[u] + 1
                    pred[v] = u
                    colour[v] = DiGraph.gray
                    Q.append(v)
            colour[u] = DiGraph.black
        return dist, pred

    def _DFS_visit(self, u): # mÃ©todo recursivo auxiliar
        self.time += 1
        self.discovered[u] = self.time
        self.colour[u] = DiGraph.gray
        for (v, weight) in self._adj[u]:
            if self.colour[v] == DiGraph.gray:
                print(f'({u},{v}) back edge')
            elif self.colour[v] == DiGraph.black:
                if self.discovered[u] < self.discovered[v]:
                    print(f'({u},{v}) forward edge')
                else:                    
                    print(f'({u},{v}) cross edge')
            elif self.colour[v] == DiGraph.white:
                print(f'({u},{v}) tree edge')
                self.pred[v] = u
                self._DFS_visit(v)
        self.colour[u] = DiGraph.black
        self.time += 1
        self.finished[u] = self.time
        
    def DFS(self):
        self.colour = { u:DiGraph.white for u in self.nodes() }
        self.pred = {}
        self.time = 0
        self.discovered = {}
        self.finished = {}
        for u in self.nodes():
            if self.colour[u] == DiGraph.white:
                self.pred[u] = None
                self._DFS_visit(u)

    def _top_sort_visit(self, u): # mÃ©todo recursivo auxiliar
        self.colour[u] = DiGraph.gray
        for (v, weight) in self._adj[u]:
            if self.colour[v] == DiGraph.gray:
                self.has_cycles = True
            elif self.colour[v] == DiGraph.white:
                self._top_sort_visit(v)
        self.colour[u] = DiGraph.black
        self.top_sorting.append(u)
        
    def topological_sorting(self): # using DFS
        self.colour = { u:DiGraph.white for u in self.nodes() }
        self.top_sorting, self.has_cycles = [], False
        for u in self.nodes():
            if self.colour[u] == DiGraph.white:
                self._top_sort_visit(u)
        self.top_sorting.reverse()
        return None if self.has_cycles else self.top_sorting
                
    def Dijkstra(self, src):
        dist = { src:0 } # asocia una distancia a cada vÃ©rtice
        pred = {} # para recuperar el camino de menor peso
        fixed = set() # cjt de vÃ©rtices que ya han sido fijados
        eligible = { src } # cjt con src como elemento inicial
        i = 0
        while len(eligible)>0:
            d,u = min( (dist[v],v) for v in eligible)
            print(f'Iter: {i} fijados: {fixed} eligible: {eligible} -> sel. {u}')
            eligible.remove(u) # lo sacamos de eligible
            fixed.add(u)       # y lo aÃ±adimos a fixed
            for v,weight in self._adj[u]: # actualizamos la cota
                if v not in fixed:        # de los adyacentes a v
                    if v not in dist:
                        pred[v], dist[v] = u, d + weight
                        eligible.add(v)
                    elif dist[v] > dist[u] + weight:
                        pred[v], dist[v] = u, d + weight
            i += 1
        return dist, pred
                
    def Dijkstra2(self, src):
        fixed = { u:False for u in self._adj }
        dist = { u:float("inf") for u in self._adj }
        dist[src] = 0
        pred = {}
        for i in range(len(self._adj)):
            d,u = min( (dist[v],v) for v in self._adj
                       if not fixed[v] )
            print(f'Iter: {i} fijados: {fixed} -> sel. {u}')
            fixed[u] = True
            for v,weight in self._adj[u]:
                if dist[v] > d + weight:
                    pred[v], dist[v] = u, d + weight
        return dist, pred
    
    def _has_cycles_visit(self,u): # mé todo recursivo auxiliar
        self.colour[u] = DiGraph.gray
        for (v,weight) in self._adj[u]:
            if self.colour[v] == DiGraph.gray: # back edge
                print(u,v)
                return True # has cycles !!!
            elif self.colour[v] == DiGraph.white: # tree edge
                if self._has_cycles_visit(v):
                    return True
        self.colour[u] = DiGraph.black
        return False

    def _has_cycles(self):
        self.colour = {u:DiGraph.white for u in self.nodes()}
        for u in self.nodes():
            if self.colour[u] == DiGraph.white:
                if self._has_cycles_visit(u):
                    return True
        return False
```

```
################################################################################################
#                                           EMPLEADOS                                          #
################################################################################################

from prettytable import PrettyTable
import collections
import datetime
import random
import names

letters = {0:"T",1:"R",2:"W",3:"A",4:"G",5:"M",6:"Y",7:"F",
           8:"P",9:"D",10:"X",11:"B",12:"N",13:"J",14:"Z",
           15:"S",16:"Q",17:"V",18:"H",19:"L",20:"C",21:"K",22:"E"}

nombres = [names.get_full_name() for i in range(200)]
cods = []
for prefix in {("cud",35),("enc",40),("rst",40),("ger",15),("seg",25),("mrg",25),("mnt",20)}:
    tipo, quantity = prefix
    for i in range(quantity):
        if len(str(i)) == 1:
            cod = f"{tipo}0{i}"
        else:
            cod = f"{tipo}{i}"
        cods.append(cod)
anyos_xp = [random.randint(0,15) for i in range(200)]
nifs = []

i = 0
while i < 200:
    number = random.randint(10000000, 99999999)
    nif = str(number) + letters[number%23]
    if nif not in nifs:
        nifs.append(nif)
        i += 1

#sueldos = [round(random.uniform(800,3000),2) for i in range(200)]
sueldos = []
for i in range(200):
    a = str(round(random.uniform(800,3000),2))
    if len(a[a.find("."):]) == 2:
        a += "0"
    a = a.replace(".",",")
    sueldos.append(a)

relations = {cod: None for cod in cods}
subs = cods.copy()
print(subs)
main_boss = subs.pop(subs.index("ger00"))
print(f"Jefe máximo: {main_boss}")
Q = collections.deque()
relations[main_boss] = main_boss
Q.append(main_boss)
while len(Q) > 0:
    u = Q.popleft()
    for i in range(random.randint(1,4)):
        if len(subs) > 0:
            sub = subs.pop(random.randint(0,len(subs)-1))
            relations[sub] = u
            #print(f"{sub} es subordinado de {u} --> Quedan {len(subs)} empleados libres")
            Q.append(sub)

g = DiGraph()
for (son,father) in relations.items():
    if son != "ger00":
        g.add_edge(father, son)
    assert(not g._has_cycles())

ini_date = datetime.date(1990,1,1)
end_date = datetime.date(2018,12,20)
diff_date = end_date - ini_date
diff_days = diff_date.days
dates = []
for i in range(200):
    r_num_days = random.randrange(diff_days)
    random_date = ini_date + datetime.timedelta(days = r_num_days)
    dates.append(random_date)

t = PrettyTable()
t.add_column("cod_emp",cods)
t.add_column("nombre",nombres)
t.add_column("Años xp", anyos_xp)
t.add_column("NIF",nifs)
t.add_column("Sueldo",sueldos)
t.add_column("Cod jefe",list(relations.values()))
t.add_column("Fecha relación",dates)

assert(len(cods)==len(nombres)==len(anyos_xp)==len(nifs)==len(sueldos)==len(list(relations.values()))==len(dates))

cad = ""
for i in range(200):
    cad += f"{cods[i]};{nombres[i]};{anyos_xp[i]};{nifs[i]};{sueldos[i]};"
    cad += f"{list(relations.values())[i]};{dates[i]}\n"

with open("Empleados.csv","w",encoding="utf-8") as f:
    for i in range(200):
        f.write(cad)



print(t.get_string(title="EMPLEADOS"))
```
